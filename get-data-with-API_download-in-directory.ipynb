```python
import urllib
import time
import feedparser
import urllib3
```

```python
# Base api query urls
base_url_arxiv = 'http://export.arxiv.org/api/query?';
```

```python
pdf_links = []

# Search parameters
search_query = 'all:cultural+neuroscience' # search for biophysics in all fields
start = 0                       # start at the first result
total_results = 400              # want 200 total results
results_per_iteration = 100       # 5 results at a time
wait_time = 3                   # number of seconds to wait beetween calls

for i in range(start,total_results,results_per_iteration):
    
    print ("Results %i - %i" % (i,i+results_per_iteration))
    
    query = 'search_query=%s&start=%i&max_results=%i' % (search_query,
                                                         i,
                                                        results_per_iteration)

    # perform a GET request using the base_url and query
    response = urllib.request.urlopen(base_url_arxiv+query).read()

    # parse the response using feedparser
    feed = feedparser.parse(response)

    # Run through each entry, and print out information
    for entry in feed.entries:
        for link in entry.links:
            if link.rel == 'alternate':
                print ('abs page link: %s' % link.href)
            elif link.title == 'pdf':
                pdf_links.append(link.href)
        #print ('arxiv-id: %s' % entry.id.split('/abs/')[-1])
        #print ('Title:  %s' % entry.title)
        # feedparser v4.1 only grabs the first author
        #print ('First Author:  %s' % entry.author)
    
    # Remember to play nice and sleep a bit before you call
    # the api again!
    print ('Sleeping for %i seconds' % wait_time) 
    time.sleep(wait_time)
```

```python
http = urllib3.PoolManager()
i = 0
for url in pdf_links:
    r = http.request('GET',url)
    Name =str(i+1) 
    i += 1
    with open("article_pdfs/file"+Name+".pdf", "wb") as f:
        f.write(r.data)
```
